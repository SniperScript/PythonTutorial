```python
class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance
    def deposit(self):
        try:
            amount = int(input("Enter the amount to deposit: "))
            if amount > 0:
                self.balance += amount
                print(f"Successfully deposited {amount}. New balance: [INR] {self.balance}")
            else:
                print("Deposit amount must be a positive number.")
        except ValueError:
            print("Invalid input. Please enter a number.")
    def withdraw(self):
        try:
            amount = int(input("Enter the amount to withdraw: "))
            if amount > self.balance:
                print("Insufficient Balance!")
            elif amount <= 0:
                print("Withdrawal amount must be a positive number.")
            else:
                self.balance -= amount
                print(f"Successfully withdrew {amount}. New balance: [INR] {self.balance}")
        except ValueError:
            print("Invalid input. Please enter a number.")
    def display(self):
        print("\n--- Account Details ---")
        print(f"Account Holder: {self.account_holder}")
        print(f"Current Balance: [INR] {self.balance}")
        print("-----------------------\n")

print("Let us create a new bank account.")
holder_name = input("Enter the account holder's name: ")

my_account = BankAccount(holder_name, 0)

my_account.display() 
my_account.deposit()    
my_account.withdraw()   
my_account.display()  
```

```python
import string
def count_words(sentence):
    normalized_sentence = sentence.lower()
    
    for char in string.punctuation:
        normalized_sentence = normalized_sentence.replace(char, ' ')
        
    words = [word for word in normalized_sentence.split() if word]
    
    word_counts = {}
    
    for word in words:
        current_count = word_counts.get(word, 0)
        word_counts[word] = current_count + 1
        
    return word_counts

text_of_wisdom = input()

frequency_map = count_words(text_of_wisdom)

word_list = [word for word in text_of_wisdom.lower().replace('.', ' ').replace('?', ' ').replace('!', ' ').split() if word]
total_words = len(word_list)
print("{")
for word, count in frequency_map.items():
    print(f" '{word}': {count}")
print("}")
```

```python
def factors(num):
    return [i for i in range(1, num + 1) if num % i == 0]

def common_factors(a, b):
    factors_a = factors(a)
    factors_b = factors(b)
    return sorted(list(set(factors_a) & set(factors_b)))

a = int(input())
b = int(input())

result = common_factors(a, b)
print(result)
```

```python
def deep_merge(d1,d2):
    out=d1.copy()
    for key,value in d2.items():
        if key in out:
            if isinstance(out[key],dict) and isinstance(value,dict):
                out[key]=deep_merge(out(key),value)
            elif isinstance(out[key],str) and isinstance(value,str):
                out[key]+=value
            elif isinstance(out[key],(int,float)) and isinstance(value,(int,float)):
                out[key]+=value
            else:
                out[key]=value
        else:
            out[key]=value
    return out
d1 = {"a": {"x": 1}, "b": 2}
d2 = {"a": {"y": 3}, "c": 4}
out=deep_merge(d1,d2)
print(out)
        
                
```

```python
def find_maximal_alternating(s):
    if not s: return []
    breaks = [i for i in range(1, len(s)) if s[i].isupper() == s[i-1].isupper()]
    indices = [0] + breaks + [len(s)]
    segments = [s[indices[i]:indices[i+1]] for i in range(len(indices)-1)]
    return [seg for seg in segments if len(seg) > 1]
s = "aAaBbCCC"
print(find_maximal_alternating(s))
```

```python
def shortest_vowel_substring(s):
    vowels = "aeiou"
    n = len(s)
    res = ""
    for i in range(n):
        found = ""
        for j in range(i, n):
            char = s[j].lower()
            if char in vowels and char not in found:
                found += char
            if len(found) >= 3:
                current = s[i:j+1]
                if not res or len(current) < len(res):
                    res = current
                break
    return res
s="earthproblem"
shortest_vowel_substring(s)
```

```python
def smallest_rotation(s):
    n = len(s)
    best_start = 0
    
    # We test every possible starting index 'i'
    for i in range(1, n):
        # Compare rotations by looking at characters one by one
        for k in range(n):
            # (i + k) % n wraps around the string like a wheel
            if s[(i + k) % n] < s[(best_start + k) % n]:
                best_start = i
                break
            elif s[(i + k) % n] > s[(best_start + k) % n]:
                break
                
    # Build the string character by character to avoid slicing
    result = ""
    for j in range(n):
        result += s[(best_start + j) % n]
    return result

s = "bac"
print(smallest_rotation(s)) # Output: "abc"
```

```python
def rotate_str(s, n):
    length = len(s)
    if length == 0: return s
    shift = n % length
    result = ""
    for i in range(length):
        index = (i + shift) % length
        result += s[index]
    return result

def smallest_rotation(s):
    n = len(s)
    if n == 0: return ""
    min_str = s
    for i in range(1, n):
        current_rotation = rotate_str(s, i)
        if current_rotation < min_str:
            min_str = current_rotation
            
    print(f"Rotated: {min_str}")

s = "bca"
smallest_rotation(s)
```

```python
import statistics
def med_freq(s):
    freqs = {c: s.count(c) for c in set(s)}
    med = statistics.median(freqs.values())
    return "".join([c for c in s if freqs[c] == med])
```

```python
def comp(d1,d2):
    diff={}
    all_keys=set(d1.keys()) | set(d2.keys())
    for k in all_keys:
        v1,v2=d1.get(k),d2.get(k)
        if isinstance(v1,dict) and isinstance(v2,dict):
            nested_diff = comp(v1,v2)
            if nested_diff:diff[k]=nested_diff
        elif v1!=v2: diff[k]=(v1,v2)
    return diff
```

```python
def matrix_ops(matrix):
    rows, cols = len(matrix), len(matrix[0])
    transpose = [[matrix[r][c] for r in range(rows)] for c in range(cols)]
    row_max = [max(r) for r in matrix]
    col_min = [min([matrix[r][c] for r in range(rows)]) for c in range(cols)]
    diag1 = sum(matrix[i][i] for i in range(min(rows, cols)))
    diag2 = sum(matrix[i][cols-1-i] for i in range(min(rows, cols)))
    return transpose, row_max, col_min, diag1 + diag2
```
