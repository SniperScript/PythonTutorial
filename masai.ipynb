```python
class car:
    def __init__(self,brand,color):
        self.brand=brand
        self.color=color
print("Enter the vehicle brand and color!")

vehicle = car(input(), input())
def display():
    print(f"Brand: {vehicle.brand},",end='')  
    print(f"Color: {vehicle.color}")
display()
```
